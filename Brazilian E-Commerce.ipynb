```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
%matplotlib inline
```

```python
orders = pd.read_csv('data/olist_orders_dataset.csv')
orders.head()
```

```python
orders.info()
```

```python
orders['order_status'].unique()
```

```python
plt.rcParams['figure.figsize'] = [12,8]
sns.countplot(orders['order_status'])
```

```python
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['day_of_year_purchase'] = orders['order_purchase_timestamp'].dt.dayofyear
orders.head()
```

```python
orders['month_purchase'] = orders['order_purchase_timestamp'].dt.month
orders.head()
```

```python
orders['day_of_week'] = orders['order_purchase_timestamp'].dt.dayofweek
orders.head()
```

```python
orders['day_of_month_purchase'] = orders['order_purchase_timestamp'].dt.day
orders.head()
```

```python
plt.figure(figsize=(18,5))
sns.distplot(orders['day_of_year_purchase'],bins=50)
```

```python
orders.drop(['order_approved_at', 
             'order_status',
             'order_delivered_carrier_date', 
             'order_delivered_customer_date', 
             'order_estimated_delivery_date'], axis=1, inplace=True)
orders.head()
```

```python
customers = pd.read_csv('data/olist_customers_dataset.csv')
customers.head()
```

```python
orders = pd.merge(orders, customers, on='customer_id')
orders.head()
```

```python
orders.drop(['customer_id','customer_unique_id'],axis=1,inplace=True)
orders.head()
```

```python
orders_items = pd.read_csv('data/olist_order_items_dataset.csv')
orders_items.head()
```

```python
orders_items = orders_items.groupby('order_id').sum().drop('order_item_id',axis=1)
orders_items.head()
```

```python
orders = pd.merge(orders, orders_items, on='order_id')
orders.head()
```

```python
orders['customer_state'].nunique()
```

```python
plt.figure(figsize=(18,5))
sns.countplot(orders['customer_state'])
```

```python
orders_pivot = orders.pivot_table(values='price',
                                  index='month_purchase',
                                  columns='day_of_month_purchase', 
                                  aggfunc=np.sum, 
                                  fill_value=0)
orders_pivot.head()
```

```python
sns.heatmap(orders_pivot,cmap='coolwarm')
```

```python
orders_sp = orders[orders['customer_state'] == 'SP']
orders_sp = orders_sp.dropna()
orders_sp.head()
```

```python
orders_pivot_sp = orders_sp.pivot_table(values='price',
                                  index='month_purchase',
                                  columns='day_of_month_purchase', 
                                  aggfunc=np.sum, 
                                  fill_value=0)
orders_pivot_sp.head()
```

```python
sns.heatmap(orders_pivot,cmap='coolwarm')
```

```python
orders_not_sp = orders[orders['customer_state'] != 'SP']
orders_not_sp = orders_not_sp.dropna()
orders_not_sp.head()
```

```python
orders_pivot_not_sp = orders_not_sp.pivot_table(values='price',
                                  index='month_purchase',
                                  columns='day_of_month_purchase', 
                                  aggfunc=np.sum, 
                                  fill_value=0)
orders_pivot_not_sp.head()
```

```python
sns.heatmap(orders_pivot_not_sp,cmap='coolwarm')
```

```python
orders_days_of_week = orders.pivot_table(values='price',
                                  index='month_purchase',
                                  columns='day_of_week', 
                                  aggfunc=np.sum, 
                                  fill_value=0)
orders_days_of_week.head()
```

```python
sns.heatmap(orders_days_of_week,cmap='coolwarm')
```
